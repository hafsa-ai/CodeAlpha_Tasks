```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"Admission_Predict.csv "  
df = pd.read_csv(file_path)
```

```python
plt.hist(df["GRE Score"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("GRE Score")
plt.ylabel("Frequency")
plt.title("Distribution of GRE Scores")
plt.show()
```

```python
sns.boxplot(x=df["CGPA"])
plt.title("Boxplot of CGPA")
plt.show()
```

```python
plt.scatter(df["GRE Score"], df["Chance of Admit "], alpha=0.5)
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admit")
plt.title("GRE Score vs Chance of Admit")
plt.show()
```

```python
sns.regplot(x="CGPA", y="Chance of Admit ", data=df, scatter_kws={"alpha":0.5})
plt.title("CGPA vs Chance of Admit")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["GRE Score", "TOEFL Score", "CGPA", "Chance of Admit "]])
plt.show()
```

```python
# data Story
# After visuals, connect them into insights:

# GRE & CGPA have strong positive correlation with admissions.

# TOEFL also matters, but less than CGPA.

# SOP/LOR strength has moderate effect.

# University Rating shows some trend, but weaker.
```
