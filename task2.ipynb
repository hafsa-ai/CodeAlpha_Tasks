```python
import pandas as pd

# Load dataset
file_path = r"Admission_Predict.csv "  
df = pd.read_csv(file_path)

# Basic info
print(df.info())
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["GRE Score"], df["Chance of Admit "])
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admit")
plt.title("GRE vs Chance of Admit")
plt.show()
```

```python
plt.scatter(df["CGPA"], df["Chance of Admit "])
plt.xlabel("CGPA")
plt.ylabel("Chance of Admit")
plt.title("CGPA vs Chance of Admit")
plt.show()
```

```python
correlation = df.corr()
print(correlation["Chance of Admit "].sort_values(ascending=False))
```

```python
import seaborn as sns

# Correlation Heatmap
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
